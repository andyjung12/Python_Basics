```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
def changeAge(x):
    if x // 10 == 0:
        return ''
    elif x // 10 == 1:
        return '10'
    elif x // 10 == 2:
        return '20'
    elif x // 10 == 3:
        return '30'
    elif x // 10 == 4:
        return '40'
    elif x // 10 == 5:
        return '50'
    elif x // 10 == 6:
        return '60'
    else:
        return ''
    
df['AgeRange'] = df['Age'].apply(changeAge)
```

```python
survived_by_ages = df.groupby('AgeRange')['Survived'].sum()
survived_by_ages
```

```python
survived_by_ages = survived_by_ages.reindex(['', '10', '20', '30', '40', '50', '60', ''])
survived_by_ages.index
```

```python
plt.bar(survived_by_ages.index, survived_by_ages.values, color='g')
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
survived_by_sex = df.groupby('Sex')['Survived'].sum()
survived_by_sex
```

```python
plt.bar(survived_by_sex.index, survived_by_sex.values, color='g')
plt.title("  ")
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
corona = pd.read_csv('./kr_daily.csv')
corona.head()
```

```python
corona.info()
```

```python
def change_date(x):
    x = str(x)
    return x[:4] + '-' + x[4:6] + '-' + x[6:]
```

```python
change_date('20200125')
```

```python
corona['date_text'] = corona['date'].apply(change_date)
corona.head()
```

```python
corona.index = corona['date_text']
corona.tail()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.title('')
plt.xticks(rotation=70)
plt.plot(corona['2020-05-18':].index, corona['2020-05-18':]['confirmed'].values, color='red', linestyle='dashed')

plt.subplot(122)
plt.title('')
plt.xticks(rotation=70)
plt.bar(corona['2020-05-18':].index, corona['2020-05-18':]['death'].values, color='red', linestyle='dashed')

plt.show()
```
