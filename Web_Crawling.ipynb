## 무신사 Web Crawling

```python
import requests
from bs4 import BeautifulSoup

for i in range(1,11): #1~10페이지
    URL = "https://search.musinsa.com/category/001".format(i)
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")
    brand_name = soup.select(".section_container_list p.item_title > a") #searchList > li:nth-child(1) > div.li_inner > div.article_info > p.item_title > a
#class 이면 앞에 . 붙히기
    item_name = soup.select(".section_container_list .list_info a") #searchList > li:nth-child(1) > div.li_inner > div.article_info > p.list_info
    price = soup.select(".section_container_list p.price") #searchList > li:nth-child(1) > div.li_inner > div.article_info > p.price

    for i in range(len(brand_name)):
        print("브랜드:", brand_name[i].text.strip())
        print("제품명:",item_name[i].attrs['title'].strip())
    
        if len(price[i].text.strip().split(" "))>=2:
            print("가격:", price[i].text.strip().split(' ')[1])
        else:
            print("가격:", price[i].text.strip())
        print(" ")
```
